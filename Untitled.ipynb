```python
import torch
from torch import nn
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pathlib
import PIL
from tqdm.auto import tqdm
```

## Load the data

```python
DATA_PATH = pathlib.Path("data/gan-getting-started/")
MONET_FILENAMES = tf.io.gfile.glob(str(DATA_PATH / "monet_tfrec" / "*.tfrec"))
print("Monet number of files:", len(MONET_FILENAMES))
PHOTO_FILENAMES = tf.io.gfile.glob(str(DATA_PATH / "photo_tfrec" / "*.tfrec"))
print("Photo number of files:", len(PHOTO_FILENAMES))
```

```python
IMAGE_SIZE = (256, 256)

def decode_image(image):
    image = tf.image.decode_jpeg(image, channels=3)
    image = (tf.cast(image, tf.float32) / 127.5) - 1
    image = tf.reshape(image, [*IMAGE_SIZE, 3])
    return image

def read_tfrecord(example):
    tfrecord_format = {
        # "image_name": tf.io.FixedLenFeature([], tf.string),
        "image": tf.io.FixedLenFeature([], tf.string), # Shape = [] means literal == "ParseSingleExample/ParseExample/ParseExampleV2:0"
        # "target": tf.io.FixedLenFeature([], tf.string)
    }
    example = tf.io.parse_single_example(example, tfrecord_format)
    image = decode_image(example["image"])
    return image

def load_dataset(filenames):
    dataset = tf.data.TFRecordDataset(filenames)
    dataset = dataset.map(read_tfrecord)
    return dataset
```

```python
class MonetPhotoDataset(torch.utils.data.Dataset):
    """
    Pytorch style dataset class. Created for processing images of Claude Monet paintings
    and common photos of nature.
    ...
    Attributes
    ----------
    images_files: list
        List of all images filenames
    transforms: torchvision.transforms.transforms.Compose
        Tranformations to apply on images data
    """
    def __init__(self, data_path, monet=True, transforms=None):
        self.monet = monet
        if monet:
            self.images_files = list(data_path.glob("monet_jpg/*.jpg"))
        else:
            self.images_files = list(data_path.glob("photo_jpg/*.jpg"))
        self.transforms = transforms
        
    def __len__(self):
        return len(self.images_files)
    
    def __getitem__(self, index):
        if self.transforms:
            result = self.transforms(self.get_image(self.images_files[index]))
        else:
            result = self.get_image(self.images_files[index])
        label = 1 if self.monet else 0
        return result, label
        
    def get_image(self, filename):
        image = PIL.Image.open(filename)
        return image
```

```python
common_transform = transforms.Compose([
    transforms.ToTensor()
])
monet_dataset = MonetPhotoDataset(DATA_PATH, transforms=common_transform)
photo_dataset = MonetPhotoDataset(DATA_PATH, monet=False, transforms=common_transform)
dataset = torch.utils.data.ConcatDataset((monet_dataset, photo_dataset))

BATCH_SIZE = 32
dataloader = torch.utils.data.DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
```

## Build the generator

```python
def downsample(in_c, out_c, size, apply_batchnorm=True):
    result = nn.Sequential()
    result.add_module(nn.Conv2d(in_c, out_c, kernel_size=size, stride=2, padding="same", bias=False))
    if apply_batchnorm:
        result.add_module(nn.BatchNorm2d(out_c))
    result.add_module(nn.LeakyReLU())
    return result

def upsample(in_c, out_c, size, apply_dropout=True):
    result = nn.Sequential()
    result.add_module(nn.ConvTranspose2d(in_c, out_c, kernel_size=size, stride=2, padding="same", bias=False))
    result.add_module(nn.BatchNorm2d(out_c))
    if apply_dropout:
        result.add_module(nn.Dropout())
    result.add_module(nn.ReLU())
    return result
```

```python
class ResidualBlock(nn.Module):
    """
    Block which represents residual connection
    with 2-x convolution layers, instance normalization and relu activation between it.
    Instance normalization and relu activation on outputs is applied.
    ...
    Attributes
    ----------
    instance_norm: torch.nn.modules.instancenorm.InstanceNorm2d
        Layer which applies normalization on each channel separately for each object in sample(batch)
    relu: torch.nn.modules.activation.ReLU
        Applies ReLU activation
    conv_block: torch.nn.modules.container.Sequential
        Block which contains convolutional1 -> instance norm -> relu -> convolutional2
    """
    def __init__(self, n_channels: int):
        super(ResidualBlock, self).__init__()
        self.instance_norm = nn.InstanceNorm2d(n_channels)
        self.relu = nn.ReLU()
        self.conv_block = nn.Sequential(*[
            nn.Conv2d(
                in_channels=n_channels,
                out_channels=n_channels,
                kernel_size=3,
                stride=1,
                padding=1
            ),
            nn.InstanceNorm2d(n_channels),
            nn.ReLU(),
            nn.Conv2d(
                in_channels=n_channels,
                out_channels=n_channels,
                kernel_size=3,
                stride=1,
                padding=1
            ),
        ])
        
    def forward(self, x: torch.Tensor):
        output = self.conv_block(x) + x
        output = self.instance_norm(output)
        output = self.relu(output)
        return output
```

```python
class Generator(nn.Module):
    """
    Convolutional autoencoder-like generator model.
    ...
    Attributes
    ----------
    reflection_pad: torch.nn.modules.padding.ReflectionPad2d
        Makes reflection pad by 3 on each side
    relu: torch.nn.modules.activation.ReLU
        Applies ReLU activation
    pixel_shuffle: torch.nn.modules.pixelshuffle.PixelShuffle
        Rearrange pixels by upscale factor == 2, output: (*, channel / factor ** 2, height * factor, width * factor)
    gathered_layers: list
        All the layers in generator model
    generator_block: torch.nn.modules.container.Sequential
        All the layers in generator model wrapped in torch.nn.Sequential instance, used in forward method
    ...
    Methods
    -------
    gather_layers:
        Makes list of all layers in generator, it is used to create generator block of model
    """
    def __init__(self,
                 in_n_channels: int,
                 hidden_n_channels: int,
                 n_residuals: int):
        super(Generator, self).__init__()
        self.reflection_pad = nn.ReflectionPad2d(3)
        self.pixel_shuffle = nn.PixelShuffle(2)
        self.relu = nn.ReLU()
        self.gathered_layers = self.gather_layers(in_n_channels, hidden_n_channels, n_residuals)
        self.generator_block = nn.Sequential(*self.gathered_layers)
        
    def forward(self, x: torch.Tensor):
        output = self.generator_block(x)
        return output
    
    def gather_layers(self,
                      in_n_channels: int,
                      hidden_n_channels: int,
                      n_residuals: int):
        layers = []
        # shape: bc * 3 * 256 * 256
        layers += [
            self.reflection_pad,
            # shape: bc * 3 * 262 * 262
            nn.Conv2d(
                in_channels=in_n_channels,
                out_channels=hidden_n_channels,
                kernel_size=7,
                stride=1,
                padding=0
            ),
            # shape: bc * hc * 256 * 256
            nn.InstanceNorm2d(hidden_n_channels),
            self.relu,
            
            nn.Conv2d(
                in_channels=hidden_n_channels,
                out_channels=2 * hidden_n_channels,
                kernel_size=3,
                stride=2,
                padding=1
            ),
            # shape: bc * 2hc * 128 * 128
            nn.InstanceNorm2d(2 * hidden_n_channels),
            self.relu,
            
            nn.Conv2d(
                in_channels=2 * hidden_n_channels,
                out_channels=4 * hidden_n_channels,
                kernel_size=3,
                stride=2,
                padding=1
            ),
            # shape: bc * 4hc * 64 * 64
            nn.InstanceNorm2d(4 * hidden_n_channels),
            self.relu,
        ]
        
        for _ in range(n_residuals):
            layers.append(ResidualBlock(4 * hidden_n_channels))
            # shape: bc * 4hc * 64 * 64
            
        layers += [
            nn.ConvTranspose2d(
                in_channels=4 * hidden_n_channels,
                out_channels=4 * 2 * hidden_n_channels,
                kernel_size=3,
                stride=1,
                padding=1),
            # shape: bc * 8hc * 64 * 64
            self.pixel_shuffle,
            # shape: bc * 2hc * 128 * 128
            nn.InstanceNorm2d(2 * hidden_n_channels),
            self.relu,
            
            nn.ConvTranspose2d(
                in_channels=2 * hidden_n_channels,
                out_channels=2 * 2 * hidden_n_channels,
                kernel_size=3,
                stride=1,
                padding=1),
            # shape: bc * 4hc * 128 * 128
            self.pixel_shuffle,
            # shape: bc * hc * 256 * 256
            nn.InstanceNorm2d(hidden_n_channels),
            self.relu,
            
            self.reflection_pad,
            # shape: bc * hc * 262 * 262
            nn.Conv2d(
                in_channels=hidden_n_channels,
                out_channels=in_n_channels,
                kernel_size=7,
                stride=1,
                padding=0),
            # shape: bc * 3 * 256 * 256
            nn.Sigmoid()
        ]
        return layers
```

```python
class Discriminator(nn.Module):
    """
    Discrimination part of the model.
    ...
    Attributes
    ----------
    main: torch.nn.modules.container.Sequential
        Sequential model with all layers
    """
    def __init__(self,
                 in_n_channels: int,
                 hidden_n_channels: int):
        super(Discriminator, self).__init__()
        self.fc = nn.Linear(in_features=30 * 30,
                            out_features=1)
        self.flattener = nn.Flatten()
        # Input shape: bc * 3 * 256 * 256
        self.main = nn.Sequential(*[
            nn.Conv2d(in_channels=in_n_channels,
                      out_channels=hidden_n_channels,
                      kernel_size=4,
                      stride=2,
                      padding=1,
                      bias=False),
            # shape: bc * hc * 128 * 128
            nn.LeakyReLU(0.2),
            
            nn.Conv2d(in_channels=hidden_n_channels,
                      out_channels=2 * hidden_n_channels,
                      kernel_size=4,
                      stride=2,
                      padding=1,
                      bias=False),
            # shape: bc * 2hc * 64 * 64
            nn.InstanceNorm2d(2 * hidden_n_channels),
            nn.LeakyReLU(0.2),
            
            nn.Conv2d(in_channels=2 * hidden_n_channels,
                      out_channels=4 * hidden_n_channels,
                      kernel_size=4,
                      stride=2,
                      padding=1,
                      bias=False),
            # shape: bc * 4hc * 32 * 32
            nn.InstanceNorm2d(4 * hidden_n_channels),
            nn.LeakyReLU(0.2),
            
            nn.Conv2d(in_channels=4 * hidden_n_channels,
                      out_channels=8 * hidden_n_channels,
                      kernel_size=4,
                      stride=1,
                      padding=1,
                      bias=True),
            # shape: bc * 8hc * 31 * 31
            nn.InstanceNorm2d(8 * hidden_n_channels),
            nn.LeakyReLU(0.2),
            
            nn.Conv2d(in_channels=8 * hidden_n_channels,
                      out_channels=1,
                      kernel_size=4,
                      stride=1,
                      padding=1,
                      bias=True)
            # shape: bc * 1 * 30 * 30
            
        ])
        
    def forward(self, x: torch.Tensor):
        output = self.main(x)
        output = self.flattener(output)
        output = self.fc(output)
        return output
```

```python
def LSGAN_D(real, fake):
    return torch.mean((real - 1) ** 2) + torch.mean(fake ** 2)

def LSGAN_G(fake):
    return torch.mean((fake - 1) ** 2)
```

```python
device = "cpu" # "mps" if torch.backends.mps.is_available() else "cpu"
generator_0 = Generator(3, 64, 6)
generator_0.to(device)
discriminator_0 = Discriminator(3, 64)
discriminator_0.to(device)
g_optimizer = torch.optim.SGD(generator_0.parameters(), lr=0.01)
d_optimizer = torch.optim.SGD(discriminator_0.parameters(), lr=0.03)
criterion = nn.BCEWithLogitsLoss()
N_EPOCHS = 5
```

```python
for epoch in tqdm(range(N_EPOCHS)):
    accum_loss = 0
    for X_batch, y_batch in tqdm(dataloader):
        X_batch, y_batch = X_batch.to(device), y_batch.to(device).to(torch.float32)
        generated_output = generator_0(X_batch)
        discriminator_logits = discriminator_0(generated_output)
        loss = criterion(discriminator_logits.flatten(), y_batch)
        g_optimizer.zero_grad()
        d_optimizer.zero_grad()
        loss.backward()
        g_optimizer.step()
        d_optimizer.step()
        accum_loss += loss.item()
    accum_loss /= len(dataloader)
    print(f"Epoch:{epoch} | Loss: {accum_loss}")
```
